#### EMISSÕES NO BRASIL

```python
# df_co2_brazil = df_co2_state.sum()
brazil_emissions = df_co2_state.sum()

plt.figure(figsize=(10,6))
plt.title(f"CO2 emissions per year in Brazil")
plt.xlabel('Year')
plt.ylabel('CO2 emissions in t')
plt.plot(brazil_emissions, label='BRASIL')
plt.legend(loc = 'best')
plt.savefig(FIG_PATH + "/BRASIL.png")
plt.show()
```

On national level, we observed a peak on CO2 emissions in 2003-2004, followed by a quick drop until 2010, when the number hit its lowest figure. Since then, we've observed a rising trend in these numbers, with a sharp rise from 2011-2013 and another one on last observed year, 2021.


```python
fig, ax = plt.subplots(figsize=(12,8))

for regiao in macrorregioes:
    tmp = df_co2_state.loc[macrorregioes[regiao]].sum()
    ax.plot(tmp, label=regiao)
ax.plot(brazil_emissions, label='Brazil')
ax.legend(loc = 'best')
ax.set_title(f"CO2 emission per region")
ax.set_xlabel('Year')
ax.set_ylabel('CO2 emissions in t')
plt.savefig(FIG_PATH + "/REGIOES.png")
plt.show()
```

Analyzing by macrorregions, we noticed that, until 2009 the North and Center-West regions varied similarly to the national numbers, being the two main contributors.
However, the same pattern is not seen in the most recent spike. In the past years, the Center-West has stalled their emissions, becoming the losing the second place of most emissions to the Southeast region for a few years, while the North has increased their participation and became the the biggest source of emission by far.

#### EMISSIONS PER STATE PER REGION

```python
def plot_per_region(df):
    for regiao in macrorregioes:
        plt.figure(figsize=(10,6))
        for estado in macrorregioes[regiao]:
            tmp = df.loc[estado]
            plt.title(f"CO2 emissions per state on {regiao} region")
            plt.xlabel('Year')
            plt.ylabel('CO2 emissions in t')
            plt.plot(tmp, label=estado)
            plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
        plt.savefig(FIG_PATH + f"/{regiao}", bbox_inches='tight')
        plt.show()
```

```python
plot_per_region(df_co2_state)
```

Whe we take a look at the states, we can make a few observations:
- Northeast: Maranhão (MA) had a sharp rise in emissions in the past year, way above all other states. In fact, the other states appear to be on a downward trend or stable.

- South: That was another region that caught our attention, given that all three states presented big rises in emissions in the past year.

- Centre-West: In this graph, we can see Brazil's former main pollutor, Mato Grosso. It is notable that, after terrible years during Brazil's peak years, in 2003 and 2004, MT has decreased its emissions by around 80% and, even though has shown a rising trend, has not increased their levels in the same rhythm as the National levels, indicating that it'is not the main one of the main contributors to the rise as it was in the beggining of the decade.

- North: Amazonas (AM) and Roraima (RR) show a positive trend in emissions in the past years. But the main problem here is Pará. It emissions are trending upwards since 2011 and, as the second biggest contributor among states, seems to be the driving force behind Brazil's worsening numbers.
